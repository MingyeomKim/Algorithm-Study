```python
from collections import deque

dx = [0, 0, 1, -1]
dy = [1, -1, 0, 0]

n, m = map(int, input().split(' '))
coin_pos = []
board = []
for row in range(n):
    line = list(input())
    coin_col = list(filter(lambda x : line[x] == 'o', range(len(line))))
    if len(coin_col) == 1:
        coin_pos.append([row, coin_col[0]])
    elif len(coin_col) == 2:
        coin_pos.append([row, coin_col[0]])
        coin_pos.append([row, coin_col[1]])
    board.append(line)
cnt_table = [[[[0 for _ in range(m)] for _ in range(n)] for _ in range(m)] for _ in range(n)]
q = deque()
q.append(coin_pos)
searched = []
searched.append(coin_pos)
res = None
cnt_table[coin_pos[0][0]][coin_pos[0][1]][coin_pos[1][0]][coin_pos[1][1]] = 1
while q:
    coin1, coin2 = q.popleft()
    i1, j1 = coin1
    i2, j2 = coin2
    if cnt_table[i1][j1][i2][j2] > 10:
        res = -1
        break
    for di, dj in zip(dx,dy):
        ni1 = i1 + di
        nj1 = j1 + dj
        ni2 = i2 + di
        nj2 = j2 + dj
        ncoin_pos = [[ni1, nj1], [ni2, nj2]]
        if (0 <= ni1 < n) and (0 <= nj1 < m) and (0 <= ni2 < n) and (0 <= nj2 < m):
            if board[ni1][nj1] == '#' and board[ni2][nj2] == '#':
                continue
            if board[ni1][nj1] == '#':
                ni1, nj1 = i1, j1
            if board[ni2][nj2] == '#':
                ni2, nj2 = i2, j2
            if not cnt_table[ni1][nj1][ni2][nj2]:
                cnt_table[ni1][nj1][ni2][nj2] = cnt_table[i1][j1][i2][j2] + 1
                q.append([[ni1,nj1],[ni2,nj2]])
        elif (ni1<0 or ni1 >=n or nj1 < 0 or nj1 >= m) and ((0 <= ni2 < n) and (0 <= nj2 < m)):
            res = cnt_table[i1][j1][i2][j2]
            break
        elif (ni2<0 or ni2 >=n or nj2 < 0 or nj2 >= m) and ((0 <= ni1 < n) and (0 <= nj1 < m)):
            res = cnt_table[i1][j1][i2][j2]
            break
            
# searched 이용하면 결과가 이상하게 나온다..
# 그리고 이 코드는 결과 출력은 잘되는데 틀렸다고 나온다..  
# 문제는 다음 탐색 칸이 보드 안에 있고 없고를 나눠서 생각하면 안된다. (벽이 있을경우 조건이 달라지기 때문에)
if res:
    print(res)
else:
    print(-1)
```
